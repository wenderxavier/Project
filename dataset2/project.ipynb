```python
#This tokenizer is used in the function before. We did not considered URL's and mentions. Later, we did not considered symbols.
def tokenizes(string, lowercase, keep_punctuation, collapse_urls, collapse_mentions):
    if not string:
        return []
    if lowercase:
        string = string.lower()
    tokens = []
    if collapse_urls:
        string = re.sub('http\S+', 'THIS_IS_A_URL', string)
    if collapse_mentions:
        string = re.sub('@\S+', 'THIS_IS_A_MENTION', string)
    if keep_punctuation:
        tokens = string.split()
    else:
        tokens = re.sub('\W+', ' ', string).split()
    return tokens
```

```python
#This method reads the file with all tweets for testing set, so the variable tokens has the data for testing
import re
tokens = []
with open("DBaecionevesAll.txt") as infile:
    for line in infile:
        tokens.append(re.sub('\W+', ' ', line).split())
```

```python
#This method takes a sample from tokens to use the same size of testing set and training set
import random
# Create feature vectors (X)
from sklearn.feature_extraction.text import CountVectorizer
random_items = []
tokens1 = []

#The range changes according to the size of the training set
#the trainging set for Aécio's data set can be 180 or 330
#the trainging set for Dilma's data set can be 250 or 429
#So, if the input is changed the range must also be changed
for i in range(1,181):
    random_items.append(random.choice(tokens))
    
#We had problems with lower case, then in this loop we do it again :)
for t in random_items:
   tokens1.append(' '.join(t).lower())
    
#Vectorizing the data    
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(t for t in tokens1)
```

```python
print 'vectorized %d tweets. found %d terms.' % (X.shape[0], X.shape[1])
```

```python
# What are the most frequent terms?
# Sum columns:
col_sums = X.sum(axis=0).tolist()[0]
# Sort sums in descending order, and return the indices.
top_indices = np.argsort(col_sums)[::-1]
print 'top_indices=', top_indices
vocab = np.array(vectorizer.get_feature_names())
top_terms = vocab[top_indices]
print 'top_terms:\n', '\n'.join('%s %d' % (term, count) for term, count in zip(top_terms, top_indices)[:10])
```

```python
# 5 Cross-validation accuracy
from sklearn.cross_validation import KFold

cv = KFold(len(y),10)
accuracies = []
for train_ind, test_ind in cv:
    model.fit(X[train_ind], y[train_ind])
    predictions = model.predict(X[test_ind])
    accuracies.append(accuracy(y[test_ind], predictions))
    
print 'Average 5-fold cross validation accuracy=%.2f (std=%.2f)' % (np.mean(accuracies), np.std(accuracies))
```
